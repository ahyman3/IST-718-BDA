### Structure of the document
    /patientId
    /path to dicom file
    /class - from class info
    /target - binary
    /age
    /sex
    /num instances
    list of target tuples (if there)
        [(x1, y1, width1, height1), (x2, y2, width2, height2),...]

```python
#Created the mongo list of dictionaries

#list of the patientIDs
patientIDs = list(set(train_labels.patientId))
#creating json for the dump
listDict = []
#looping through the patient IDs
for patient in patientIDs:
    patientDict = {}
    patientDict["patientID"] = patient
    #File name for patient
    filename = "/Users/alexhyman/RSNA/stage_2_train_images/" + patient + ".dcm"
    #Adding the file path
    patientDict["path"] = filename
    #read patient file
    filecontent = pyd.read_file(filename)
    #Adding sex to json
    patientDict["sex"] = filecontent.PatientSex
    #Adding age to json
    patientDict["age"] = filecontent.PatientAge    
    #indexing the patient in the training df
    patient_xray = train_labels.loc[train_labels.patientId == patient, ]
    #Indexing the xray info
    xray_info = classinfo.loc[classinfo.patientId == patient, ]
    #Getting class of the xray
    xray_class = xray_info.iloc[0, 1]
    patientDict["class"] = xray_class
    #getting if target
    if patient_xray.shape[0] > 1 or patient_xray.Target.values[0] == 1:
        target = 1
    else:
        target = 0
    #if target is 1
    if target == 1:
        targets = []
        for index, row in patient_xray.iterrows():
            instance = (row['x'], row["y"], row['width'], row['height'])
            targets.append(instance)
        patientDict["targets"] = targets
        patientDict["num_instances"] = len(targets)
    else:
        patientDict["num_instances"] = 0
    patientDict["target"] = target
    listDict.append(patientDict)
```

```python
#Inserting the list of documents in the mongodb
col.insert_many(listDict)
```
